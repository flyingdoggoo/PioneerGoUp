```python
from nanodet.util import overlay_bbox_cv

from IPython.display import display
from PIL import Image

def cv2_imshow(a, convert_bgr_to_rgb=True):
    """A replacement for cv2.imshow() for use in Jupyter notebooks.
    Args:
        a: np.ndarray. shape (N, M) or (N, M, 1) is an NxM grayscale image. shape
            (N, M, 3) is an NxM BGR color image. shape (N, M, 4) is an NxM BGRA color
            image.
        convert_bgr_to_rgb: switch to convert BGR to RGB channel.
    """
    a = a.clip(0, 255).astype('uint8')
    # cv2 stores colors as BGR; convert to RGB
    if convert_bgr_to_rgb and a.ndim == 3:
        if a.shape[2] == 4:
            a = cv2.cvtColor(a, cv2.COLOR_BGRA2RGBA)
        else:
            a = cv2.cvtColor(a, cv2.COLOR_BGR2RGB)
    display(Image.fromarray(a))
```

```python
import cv2
import time
from nanodet.util import overlay_bbox_cv
from demo.demo import Predictor
from nanodet.util import cfg, load_config, Logger # đảm bảo bạn import đầy đủ các hàm cần thiết
import torch

# Giả sử bạn đã có file config, model checkpoint và thiết lập device
cfg_path = "/content/nanodet_clone/config/nanodet-plus-m_416-yolo.yml"
model_path = "/content/nanodet_clone/workspace/nanodet-plus-m_416/model_best/nanodet_model_best.pth"
device = torch.device('cpu')

# Load config và tạo predictor
load_config(cfg, cfg_path)
logger = Logger(-1, use_tensorboard=False)
predictor = Predictor(cfg, model_path, logger, device=device)

# Mở video (đường dẫn tới file video hoặc sử dụng camera bằng chỉ số 0)
video_path = "/content/video_demo.mp4"  # hoặc sử dụng 0 cho webcam
cap = cv2.VideoCapture(video_path)
fps = cap.get(cv2.CAP_PROP_FPS)
w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
h = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
out = cv2.VideoWriter('output_video.mp4', cv2.VideoWriter_fourcc(*'mp4v'), fps, (w, h))
total_frame = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

if not cap.isOpened():
    print("Không thể mở video:", video_path)
    exit()

start_time = time.time()

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Bắt đầu đo thời gian xử lý cho frame này

    # Thực hiện dự đoán trên frame
    # Lưu ý: predictor.inference có thể được thiết kế để nhận đường dẫn hoặc ảnh (np.array)
    # Nếu code gốc chỉ nhận image_path, bạn cần chỉnh sửa predictor để hỗ trợ np.array (hoặc lưu frame tạm thời)
    meta, res = predictor.inference(frame)

    # Áp dụng overlay bounding box lên ảnh gốc.
    # Trong code mẫu ban đầu, meta['raw_img'][0] chứa ảnh gốc.
    result_frame = overlay_bbox_cv(meta['raw_img'][0], res[0], cfg.class_names, score_thresh=0.35)
    # out.write(result_frame)
    # Kết thúc đo thời gian và tính FPS


end_time = time.time()

fps = total_frame / (end_time - start_time)
print('\n')
print("fps: ",fps)

out.release()
cap.release()
```

```python
import cv2
import time
from ultralytics import YOLO

model = YOLO("/content/crack_yolov11n_fix.pt")  # Thay đổi đường dẫn nếu bạn sử dụng model khác hoặc model tự huấn luyện
model.to('cpu')
video_path = "/content/video_demo.mp4"  # hoặc sử dụng 0 cho webcam
cap = cv2.VideoCapture(video_path)
fps = cap.get(cv2.CAP_PROP_FPS)
w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
h = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
out = cv2.VideoWriter('output_video.mp4', cv2.VideoWriter_fourcc(*'mp4v'), fps, (w, h))
total_frame = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

if not cap.isOpened():
    print("Không thể mở video:", video_path)
    exit()

start_time = time.time()

while True:
    ret, frame = cap.read()
    if not ret:
        break
    results = model.predict(frame, verbose=False)

    # 5. Lấy ảnh đã được vẽ bounding box (annotated image)
    # results là danh sách các đối tượng kết quả, với mỗi phần tử có phương thức plot() trả về ảnh đã được annotate.
    annotated_frame = results[0].plot()  # annotated_frame là ảnh kiểu numpy
    # out.write(annotated_frame)

end_time = time.time()
fps = total_frame / (end_time - start_time)
print("fps: ",fps)
out.release()
cap.release()
```
